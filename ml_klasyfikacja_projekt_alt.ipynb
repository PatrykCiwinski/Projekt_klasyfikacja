```python
import seaborn as sns

```

```python
y=water.pop('Potability')
```

```python
X=water
```

#### Dyskretyzacja zmiennych

```python
from sklearn.preprocessing import KBinsDiscretizer
```

```python
disc=KBinsDiscretizer(n_bins=5, # liczba przedziałów
                      encode='ordinal', # wyświetlanie tabeli w rzeczywistym formacie
                      strategy='uniform') # tworzenie przedziałów, aby miały taką samą szerokość przedziału wartości
```

#### SVC

```python
y
```

```python
X_disc=disc.fit_transform(X)
```

```python
X_disc
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train, y_test=train_test_split(X_disc, y, test_size=0.2, random_state=123)
```

```python
scaler = StandardScaler()
```

```python
X_train_scaler = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC


svm=SVC(kernel='poly', class_weight='balanced')
svm.fit(X_train_scaler,y_train)
pred=svm.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, pred))
```

```python
cm=confusion_matrix(y_test, pred)
sns.heatmap(cm, annot=True, fmt='d')
```

```python
svm1=SVC(kernel='linear')
svm1.fit(X_train_scaler,y_train)
pred1=svm1.predict(X_test)
```

```python
print(classification_report(y_test, pred1))
```

```python
cm1=confusion_matrix(y_test, pred1)
sns.heatmap(cm1, annot=True, fmt='d')
```

#### GridSearch

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters={'C':[0.01, 0.1, 1, 10, 100], 'gamma':['scale', 'auto']}
```

```python
svm=SVC(kernel='poly')
```

```python
searcher=GridSearchCV(svm, parameters)
searcher.fit(X_train_scaler,y_train)
```

```python
print("Best C parameter", searcher.best_params_)
```

```python
print("Best score", searcher.best_score_)
```

```python
searcher_pred=searcher.predict(X_test)
```

```python
print(classification_report(y_test, searcher_pred))
```

```python
cm_search=confusion_matrix(y_test, searcher_pred)
sns.heatmap(cm_search, annot=True, fmt='d')
```

#### KNN- algorytm najbliższych sąsiadów

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_disc, y, test_size=0.2)
```

```python
knn=KNeighborsClassifier(n_neighbors=5, metric='minkowski')
```

```python
knn.fit(X_train, y_train)
knn_pred=knn.predict(X_test)
```

```python
print(classification_report(y_test, knn_pred))
```

```python
cm_knn=confusion_matrix(y_test, knn_pred)
sns.heatmap(cm_knn, annot=True, fmt='d')
```

```python
knn=KNeighborsClassifier()
```

```python
from sklearn.pipeline import Pipeline
```

```python
# operations = [('scaler',StandardScaler()),('knn',knn)]
knn2 = Pipeline([('scaler',StandardScaler()),('knn',knn)])
```

```python
k_values = list(range(1,30))
```

```python
# definiowanie parametrów do testowania KNN
params_knn={'knn__metric':['minkowski', 'chebyshev', 'manhattan', 'euclidean'],'knn__n_neighbors':k_values}
```

```python
knn_classify=GridSearchCV(knn2, params_knn, cv=5)
```

```python
knn_classify.fit(X_train, y_train)
```

```python
knn_classify.best_estimator_.get_params()
```

```python
knn_classify.best_score_
```

```python
knn_pred2 = knn_classify.predict(X_test)
```

```python
print(classification_report(y_test, knn_pred2))
```

```python
cm_knn2=confusion_matrix(y_test, knn_pred2)
sns.heatmap(cm_knn2, annot=True, fmt='d')
```
